```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/marketing_campaign.csv", sep="\t")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.boxplot(y=df["Income"])
plt.title("Income Distribution")
plt.savefig("../outputs/income_boxplot.png")
plt.show()
```

```python
sns.barplot(x=df["Education"], y=df["Income"])
plt.title("Income by Education Level")
plt.savefig("../outputs/education_income_bar.png")
plt.show()
```

Observation:
Customers with higher income levels show greater spending potential.
Education level influences income and purchasing behavior, which can guide
targeted marketing strategies in banking.
